```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from pathlib import Path
from sklearn.preprocessing import StandardScaler
```

## Preparing the data for the model

```python
 # Loading data
    
file_path = Path("Resources/myopia.csv")
df_myopia = pd.read_csv(file_path)
df_myopia.head()
```

```python
# Check data types

df_myopia.dtypes
```

```python
# check for duplicates

df_myopia.duplicated().sum()
```

```python
# Drop Myopic column
labels = df_myopia['MYOPIC']
df_myopia = df_myopia.drop(columns='MYOPIC', axis = 1)
df_myopia
```

```python
# Scale the data

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_myopia)
```

```python
# Create a DataFrame with the transformed data

df_myopia_scaled = pd.DataFrame(scaled_data, columns=df_myopia.columns)

df_myopia_scaled.head()
```

## Dimension reduction using PCA

```python
from sklearn.decomposition import PCA
```

```python
# Initialize PCA model

pca = PCA(n_components=0.9)

# Get two principal components for the myopia data.
myopia_pca = pca.fit_transform(df_myopia_scaled)
```

```python
# Transform PCA data to a DataFrame

df_myopia_pca = pd.DataFrame(data=myopia_pca)
df_myopia_pca.head()
```

## Further dimension reduction using t-SNE

```python
from sklearn.manifold import TSNE
```

```python
# Initialize t-SNE model
tsne = TSNE(learning_rate=80)

 # Reduce dimensions
tsne_features = tsne.fit_transform(df_myopia_pca)
```

```python
tsne_features.shape
```

```python
 # Visualize the clusters
plt.scatter(tsne_features[:,0], tsne_features[:,1], c=labels)
plt.show()
```

## Cluster Analysis with K-means

```python
# Finding the best value for k
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(df_myopia_pca)
    inertia.append(km.inertia_)

# Creating the Elbow Curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)

plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(list(range(11)))
plt.title('Elbow Curve')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
